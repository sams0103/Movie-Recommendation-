```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv("ratings.csv")
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("movies.csv")
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='movieId')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='num of ratings',data=ratings,alpha=0.5)
```

```python
moviemat = df.pivot_table(index='userId',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
ForestGump_user_ratings = moviemat['Forrest Gump (1994)']
MatrixThe_user_ratings = moviemat['Matrix, The (1999)']
ForestGump_user_ratings.head()
```

```python
similar_to_ForestGump = moviemat.corrwith(ForestGump_user_ratings)
similar_to_MatrixThe = moviemat.corrwith(MatrixThe_user_ratings)
```

```python
corr_ForestGump = pd.DataFrame(similar_to_ForestGump,columns=['Correlation'])
corr_ForestGump.dropna(inplace=True)
corr_ForestGump.head()
```

```python

corr_ForestGump.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_ForestGump = corr_ForestGump.join(ratings['num of ratings'])
corr_ForestGump.head()
```

```python
corr_ForestGump[corr_ForestGump['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_MatrixThe = pd.DataFrame(similar_to_MatrixThe,columns=['Correlation'])
corr_MatrixThe.dropna(inplace=True)
corr_MatrixThe = corr_MatrixThe.join(ratings['num of ratings'])
corr_MatrixThe[corr_MatrixThe['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
